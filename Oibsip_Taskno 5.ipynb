```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
```

```python
raw_data = pd.read_csv('Dataset/Advertising.csv')
raw_data
```

```python
raw_data = raw_data.drop('Unnamed: 0',axis=1)
```

```python
raw_data
```

```python
raw_data.describe()
```

```python
X = raw_data.drop('Sales',axis = 1)
y = raw_data['Sales']
X,y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state=555)
```

```python
model = RandomForestRegressor(n_estimators = 100, random_state = 20)
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
predictions
```

```python
r2_score = r2_score(y_test,predictions)
r2_score
```

```python
mse = mean_squared_error(y_test,predictions)
mse
```
