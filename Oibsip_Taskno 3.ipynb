```python
data['Fuel_Type'].unique(),data['Selling_type'].unique(),data['Transmission'].unique()
```

```python
data['Fuel_Type'] = data['Fuel_Type'].replace({'Petrol' : 0, 'Diesel' : 1, 'CNG' : 2})
data['Selling_type'] = data['Selling_type'].replace({'Dealer' : 0, 'Individual' : 1})
data['Transmission'] = data['Transmission'].replace({'Manual' : 0, 'Automatic' : 1})
```

```python
data
```

```python
current_year = 2025
data['Car_age'] = current_year - data['Year']
```

```python
data
```

```python
data.drop(['Car_Name','Year'],inplace = True,axis = 1)
```

```python
data
```

```python
feature = ['Present_Price','Driven_kms','Fuel_Type','Selling_type','Transmission','Owner','Car_age']
feature
```

```python
data[feature]
```

```python
X = data[feature]
y = data['Selling_Price']
X,y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2,random_state = 50)
```

```python
model = RandomForestRegressor(n_estimators = 100, random_state = 50)
```

```python
model.fit(X_train,y_train)
```

```python
y_prediction = model.predict(X_test)
```

```python
y_prediction
```

```python
model_r2_score = r2_score(y_test,y_prediction)
```
